```python
from pathlib import Path
from scipy.stats import gaussian_kde 

import pickle
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

plt.rcParams["mathtext.fontset"] = "cm" 
plt.rcParams["font.family"] = "serif"
```

```python
all_rc = Path("/Users/vinland/mulab/main/assets/ALL_RC.pickle") 
with open(all_rc, "rb") as f: 
    catalog = pickle.load(f)

mF212N = catalog["mF212N"]
mmin, mmax = mF212N.min(), mF212N.max() 
```

```python
def generate_bins(mmin: float, mmax: float, n: int):
    vals = np.linspace(mmin, mmax, n + 1)
    return vals  

def generate_catalog(catalog, values, mmin, mmax, n, min_size=1000):
    edges = generate_bins(mmin, mmax, n)
    N = len(values)

    if N <= min_size:
        return {0: catalog.copy()}, np.array([(mmin, mmax)])

    masks = []
    merged_edges = []

    acc_mask = np.zeros(N, dtype=bool)
    acc_count = 0
    start_edge = edges[0]

    for i in range(n):
        left, right = edges[i], edges[i+1]
        if i < n - 1:
            mask = (values >= left) & (values < right)
        else:
            mask = (values >= left) & (values <= right)

        acc_mask |= mask
        acc_count += int(mask.sum())

        if acc_count >= min_size:
            masks.append(acc_mask)
            merged_edges.append((start_edge, right))
            acc_mask = np.zeros(N, dtype=bool)
            acc_count = 0
            start_edge = right

    if acc_count > 0:
        if masks:
            masks[-1] = masks[-1] | acc_mask
            s, _ = merged_edges[-1]
            merged_edges[-1] = (s, edges[-1])
        else:
            masks.append(acc_mask)
            merged_edges.append((edges[0], edges[-1]))

    catalogs = {i: catalog[m] for i, m in enumerate(masks)}
    return catalogs, np.array(merged_edges)
```

```python
n = 6
min_size = 400

t, final_bins = generate_catalog(
    catalog=catalog,
    values=mF212N,   
    mmin=15.3, 
    mmax=15.625,
    n=n,
    min_size=min_size
)

with open(f"pickle/catalog_n={len(t)}_{min_size=}.pickle", "wb") as f: 
    pickle.dump(t, f)

for m in t.values(): 
    print(len(m))
```

```python
slope_path = Path("/Users/vinland/mulab/main/mag_variance/pickle/") 

with open(slope_path / "slope_dict_6bins.pickle", "rb") as f: 
    E_A = pickle.load(f)
```

```python
# (AF115W - AF212N)/AF212N 
E_A
```

```python
ratios = [] 
errors = [] 
median = []

for idx, (slope, error, intercept) in E_A.items(): 
    ratios.append(1 + slope)
    errors.append(error) 
    median.append(np.mean(final_bins[idx]))
```

```python
fig, axis = plt.subplots(1, 1, figsize=(8, 6))

plt.errorbar(
    median, ratios, yerr=errors,
    fmt='d',                     
    color='black',
    ecolor='gray',              
    elinewidth=1.2,
    capsize=3,                   
    markersize=5,
    label=r"$A_{F115W}/A_{F212N}$"
)

plt.xlabel(r"$m_{F212N}$ (median bin center)", fontsize=15)
plt.ylabel(r"$A_{F115W}/A_{F212N}$", fontsize=15)
plt.title(r"$A_{F115W}/A_{F212N}$ vs. $A_{F212N}$ (Proxy)", fontsize=15)
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()

#plt.savefig("media/AF115W_AF212N_vsAF212N.png", dpi=300)
```

```python
from scipy.stats import gaussian_kde

m1 = [] 
m2 = []

for s in t.values(): 
    m115 = s["mF115W"] 
    m212 = s["mF212N"] 

    for m in m115: 
        m1.append(m) 
    for m in m212: 
        m2.append(m) 

x = np.asarray(m2) 
y = np.subtract(m1, m2)

xy = np.vstack([x, y]) 
z = gaussian_kde(xy)(xy) 

fig, axis = plt.subplots(1, 1, figsize=(8, 6)) 

plt.scatter(x, y, c=z, s=1)
plt.xlabel(r'$m_{F212N}$', fontsize=15)
plt.ylabel(r'$m_{F115W} - m_{F212N}$', fontsize=15)
```
