```python
import pickle 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

plt.rcParams['font.family'] = 'serif' 
plt.rcParams['font.size']   = 14
plt.rcParams['mathtext.fontset'] = 'cm'

from pathlib import Path 
from scipy.stats import gaussian_kde 
```

```python
pickle_dir = Path.home() / "mulab/main/assets/" 
nrcb1  = pickle_dir / "NRCB1.pickle"
nrcb2  = pickle_dir / "NRCB2.pickle"
nrcb3  = pickle_dir / "NRCB3.pickle"
nrcb4  = pickle_dir / "NRCB4.pickle"

with open(nrcb1, "rb") as f: nb1 = pickle.load(f)
with open(nrcb2, "rb") as f: nb2 = pickle.load(f)
with open(nrcb3, "rb") as f: nb3 = pickle.load(f)
with open(nrcb4, "rb") as f: nb4 = pickle.load(f)

# full RC dataset 
nb = pd.concat([nb1, nb2, nb3, nb4], ignore_index=True)
```

```python
# magnitudes 
mf115w1 = np.asarray(nb1["mF115W"], dtype=float) 
mf212n1 = np.asarray(nb1["mF212N"], dtype=float) 
mf323n1 = np.asarray(nb1["mF323N"], dtype=float)
mf405n1 = np.asarray(nb1["mF405N"], dtype=float)

mf115w2 = np.asarray(nb2["mF115W"], dtype=float) 
mf212n2 = np.asarray(nb2["mF212N"], dtype=float) 
mf323n2 = np.asarray(nb2["mF323N"], dtype=float)
mf405n2 = np.asarray(nb2["mF405N"], dtype=float)

mf115w3 = np.asarray(nb3["mF115W"], dtype=float) 
mf212n3 = np.asarray(nb3["mF212N"], dtype=float) 
mf323n3 = np.asarray(nb3["mF323N"], dtype=float)
mf405n3 = np.asarray(nb3["mF405N"], dtype=float)

mf115w4 = np.asarray(nb4["mF115W"], dtype=float) 
mf212n4 = np.asarray(nb4["mF212N"], dtype=float) 
mf323n4 = np.asarray(nb4["mF323N"], dtype=float)
mf405n4 = np.asarray(nb4["mF405N"], dtype=float)

# coords 
mra1 = np.asarray(nb1["ra"],  dtype=float)
mde1 = np.asarray(nb1["dec"], dtype=float)

mra2 = np.asarray(nb2["ra"],  dtype=float)
mde2 = np.asarray(nb2["dec"], dtype=float)

mra3 = np.asarray(nb3["ra"],  dtype=float)
mde3 = np.asarray(nb3["dec"], dtype=float)

mra4 = np.asarray(nb4["ra"],  dtype=float)
mde4 = np.asarray(nb4["dec"], dtype=float)

# const ra/dec of sagA* SMBH
SAG_A = (266.41683, -29.00781) 
```

```python
coords1 = np.stack([mra1, mde1], axis=1)
coords2 = np.stack([mra2, mde2], axis=1)
coords3 = np.stack([mra3, mde3], axis=1)
coords4 = np.stack([mra4, mde4], axis=1)

# array of distances to sag A* 
radii1 = np.hypot(
    coords1[:,0] - SAG_A[0],
    coords1[:,1] - SAG_A[1]
)
radii2 = np.hypot(
    coords2[:,0] - SAG_A[0],
    coords2[:,1] - SAG_A[1]
)
radii3 = np.hypot(
    coords3[:,0] - SAG_A[0],
    coords3[:,1] - SAG_A[1]
)
radii4 = np.hypot(
    coords4[:,0] - SAG_A[0],
    coords4[:,1] - SAG_A[1]
)
```

```python
# proxy plot against F323N
x1 = radii1
y1 = (mf323n1 - mf212n1)/(mf115w1 - mf212n1)

x2 = radii2
y2 = (mf323n2 - mf212n2)/(mf115w2 - mf212n2)

x3 = radii3
y3 = (mf323n3 - mf212n3)/(mf115w3 - mf212n3)

x4 = radii4
y4 = (mf323n4 - mf212n4)/(mf115w4 - mf212n4)


mask1 = np.isfinite(x1) & np.isfinite(y1) 
mask2 = np.isfinite(x2) & np.isfinite(y2) 
mask3 = np.isfinite(x3) & np.isfinite(y3) 
mask4 = np.isfinite(x4) & np.isfinite(y4) 

x1 = x1[mask1] 
y1 = y1[mask1]
x2 = x2[mask2] 
y2 = y2[mask2]
x3 = x3[mask3] 
y3 = y3[mask3]
x4 = x4[mask4] 
y4 = y4[mask4]
```

```python
fig, axis = plt.subplots(1, 1, figsize=(8, 8)) 

xy1 = np.vstack([x1, y1]) 
z1  = gaussian_kde(xy1)(xy1) 
xy2 = np.vstack([x2, y2]) 
z2  = gaussian_kde(xy1)(xy2) 
xy3 = np.vstack([x3, y3]) 
z3  = gaussian_kde(xy1)(xy3) 
xy4 = np.vstack([x4, y4]) 
z4  = gaussian_kde(xy1)(xy4) 

plt.scatter(x1*3600, y1+0.3, c=z1, cmap='magma',   s=3, label="NRCB1")
plt.scatter(x2*3600, y2+0.15, c=z2, cmap='viridis', s=3, label="NRCB2")
#plt.scatter(x3*3600, y3, c=z3, cmap='plasma',  s=3, label="NRCB3")
plt.scatter(x4*3600, y4, c=z4, cmap='cividis', s=3, label="NRCB4")

plt.xlabel(r"angular distance from Sag $A^*$ (arcsec)", fontsize=20)
plt.ylabel(r"$(m_{F323N} - m_{F212N}) / (m_{F115W} - m_{F212N})$", fontsize=20)
plt.legend(markerscale=5)
```

```python
# just combining all the NRCB data together 
x = np.concatenate([x1, x2, x3, x4])
y = np.concatenate([y1, y2, y3, y4])

xy = np.vstack([x, y]) 
z  = gaussian_kde(xy)(xy) 
```

```python
fig, axis = plt.subplots(1, 1, figsize=(8, 8)) 

plt.scatter(x*3600, y, c=z, s=2, cmap='magma', label="All NRCBs")
plt.xlabel("angular distance from Sag A* (arcsec)", fontsize=20) 
plt.ylabel(r"$(m_{F323N} - m_{F212N}) / (m_{F115W} - m_{F212N})$", fontsize=20)
plt.legend(markerscale=5)
plt.savefig("raw/sagA_F323N_individual.png", dpi=300)
```
