```python
from astropy.io import fits
import matplotlib.pyplot as plt
from pathlib import Path

fits_file = Path.home() / "desktop/jw01939-o001_t001_nircam_clear-f212n_i2d.fits"

# Open FITS file
hdul = fits.open(fits_file)
hdul.info()
```

```python
from astropy.wcs import WCS
from astropy.visualization import ImageNormalize, ZScaleInterval, AsinhStretch

with fits.open(fits_file, memmap=True) as hdul:
    data = np.nan_to_num(hdul['VAR_POISSON'].data, nan=0.0, posinf=0.0, neginf=0.0)
    wcs = WCS(hdul['SCI'].header, hdul).celestial  

norm = ImageNormalize(data, interval=ZScaleInterval(), stretch=AsinhStretch())

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection=wcs)
im = ax.imshow(data, origin='lower', cmap='cubehelix', norm=norm)

lon = ax.coords[0] 
lat = ax.coords[1]

lon.set_format_unit(u.deg)         
lat.set_format_unit(u.deg)

lon.set_major_formatter('d.ddd')   
lat.set_major_formatter('d.ddd')

ax.set_xlabel('RA [deg]')
ax.set_ylabel('Dec [deg]')
ax.coords.grid(True, color='white', ls=':', alpha=0.4)

fig.colorbar(im, ax=ax, label='flux')
plt.tight_layout()
plt.show()
```

```python
import astropy.units as u
import numpy as np

with fits.open(fits_file, memmap=True) as hdul:
    var = hdul['VAR_POISSON'].data
    w   = WCS(hdul['SCI'].header, hdul).celestial

valid    = np.isfinite(var) & (var > 0)
row_frac = valid.mean(axis=1)
mask_gap = row_frac < 0.02  

edges = np.diff(np.r_[False, mask_gap, False]).nonzero()[0]
runs = list(zip(edges[::2], edges[1::2]))          
gap_start, gap_stop = max(runs, key=lambda ab: ab[1] - ab[0])
y_split = (gap_start + gap_stop) // 2              


# access top half (NRCB) 
y0, y1 = y_split, var.shape[0]                      
var_top = var[y0:y1, :]
w_top = w.slice((slice(y0, y1), slice(None)))       

norm = ImageNormalize(var_top, interval=ZScaleInterval(), stretch=AsinhStretch())

fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(111, projection=w_top)
im = ax.imshow(np.nan_to_num(var_top), origin='lower', cmap='cubehelix', norm=norm)

lon, lat = ax.coords
lon.set_format_unit(u.deg); lat.set_format_unit(u.deg)
lon.set_major_formatter('d.ddd'); lat.set_major_formatter('d.ddd')
ax.coords.grid(True, color='white', ls=':', alpha=0.4)

ax.set_xlabel('RA [deg]')
ax.set_ylabel('Dec [deg]')
fig.colorbar(im, ax=ax, label='flux')
plt.gca().invert_xaxis()
plt.tight_layout(); plt.show()
```
